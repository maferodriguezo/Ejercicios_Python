#**Escribe un bucle <code>for</code> que imprima todos los elementos entre <b>-5</b> y <b>5</b> usando la función range.**

```python
for i in range (-5,6):
  print(i)
```

#**Escribe un bucle for que imprima la siguiente lista: <code>squares=['red', 'yellow', 'green', 'purple', 'blue']</code>**

```python
squares=['red', 'yellow', 'green', 'purple', 'blue']

N = len(squares)

for i in range(N):
  print(squares[i])
```

**Respuesta:** Se hizo uso de "N = len(squares)" para tener el tamaño de la lista y luego se usa en el bucle ese limite para mostrar los elementos

#**Escribe un bucle while que muestre los valores de la calificación de las canciones de un disco almacenados en la lista <code>PlayListRatings</code>. Si la calificación es menor a 6, sal del bucle. La lista <code>PlayListRatings</code> esta compuesta por: <code>PlayListRatings = [10, 9.5, 10, 8, 7.5, 5, 10, 10]</code>**

```python
PlayListRatings = [10, 9.5, 10, 8, 7.5, 5, 10, 10]

i = 0
calificacion = 0

while(calificacion >= 6 or i == 0):
    calificacion = PlayListRatings[i]
    i += 1
    print(calificacion)
```

**Respuesta:** Se hizo uso de condicionales para que el bucle salga al momento de encontrar un nuemro que no sea mayor o igual a 6, por lo tanto no muestra más numeros despues de leer la calificacion con 5

#**Escribe un bucle while que copie las cadenas <code>'orange'</code> de la lista <code>squares</code> a la lista <code>new_squares</code>. Detente y sal del ciclo si el valor en la lista no es <code>'orange'</code>:**

```python
squares = ['orange', 'orange', 'purple', 'blue ', 'orange']
new_squares = []

i = 0

while squares[i] == 'orange':
    new_squares.append(squares[i])  # Añadimos el elemento a new_squares
    i += 1

print("Lista new_squares:", new_squares)
```

**Respuesta:** Se hizo uso de "append", para que se agregue a la lista y tambien "while squares[i] == 'orange' ", para que itere y guarde los elementos de la lista con ese nombre
