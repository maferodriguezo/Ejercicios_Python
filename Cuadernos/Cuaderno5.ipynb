#**<b>Necesitarás este diccionario para las siguientes dos preguntas:</b>**

```python
# Pregunta - Diccionario de ejemplo

soundtrack_dic = {"The Bodyguard":"1992", "Saturday Night Fever":"1977"}
soundtrack_dic
```

**a) En el diccionario <code>soundtrack_dict</code>. ¿Cuáles son las llaves?**

**Respuesta:** A simple vista se identifica que las llaves son "The Bodyguard" y "Saturday Night Fever". Aun asi usaré el llamado de llaves para verificar con ayuda de "soundtrack_dic.keys"

```python
soundtrack_dic.keys()
```

**b) En el diccionario <code>soundtrack_dict</code>. ¿Cuáles son los valores?**

**Respuesta:** A simple vista se identifica que los valores son "1992" y "1977". Aun asi usaré el llamado de valores para verificar con ayuda de "soundtrack_dic.values"

```python
soundtrack_dic.values()
```

#**<b>Necesitarás este diccionario para las siguientes preguntas:</b>**

Los discos <b>Back in Black</b>, <b>The Bodyguard</b> y <b>Thriller</b> poseen el siguiente record de ventas en millones, 50, 50 y 65 respectivamente

```python
discos = { "Back in Black": 50, "The Bodyguard": 50, "Thriller": 65}
discos
```

**b) Usa el diccionario para encontrar las ventas totales de <b>Thriller</b>:**

```python
discos['Thriller']
```

**Respuesta:** Se hizo uso de "discos['Thriller']" lo cual nos mostrara las ventas totales de ese disco que son "65"

**c) Encuentra los nombres de los discos dentro del diccionario mediante el método <code>keys</code>:**

```python
discos.keys()
```

**Respuesta:** Se hizo uso de "discos.keys()", dando como respuesta "(['Back in Black', 'The Bodyguard', 'Thriller'])"

**d) Encuentra los records de ventas dentro del diccionario mediante el método <code>values</code>:**

```python
discos.values()
```

**Respuesta:** Se hizo uso de "discos.values()", dando como respuesta "([50, 50, 65])"
