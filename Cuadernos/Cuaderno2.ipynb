#**Solución de examen sobre tuplas**

**Considérese la siguiente tupla**

```python
# Tupla de ejemplo

genres_tuple = ("pop", "rock", "soul", "hard rock", "soft rock", \
                "R&B", "progressive rock", "disco")
genres_tuple
```

#**Encontrar la longitud de la tupla, <code>genres_tuple</code>:**

```python
len(genres_tuple)
```

**Respuesta:** Se hizo uso de "len(genres_tuple)", lo cual nos dio a conocer que la longitud de la tupla es de 8

#**Accede al elemento en el índice 3:**

```python
elemento = genres_tuple[3]

print(elemento)
```

**Respuesta:** Se hizo uso de "genres_tuple[3]" y se guardo en una variable, lo cual nos muestra que el elemento en el indice 3 es "hard rock"

#**Utiliza el slicing para obtener los índices 3, 4 y 5**

```python
slicing = genres_tuple[3:6]
print(slicing)
```

**Respuesta:** Se hizo uso de "genres_tuple[3:6]", lo cual nos dio a conocer los indices del 3 al 5 "'hard rock', 'soft rock', 'R&B'"

#**Encuentra los primeros dos elementos de la tupla <code>genres_tuple</code>**

```python
encontrar_elementos = genres_tuple[0:2]
print(encontrar_elementos)
```

**Respuesta:** Se hizo uso de "genres_tuple[0:2]", como el punto anterior, pero aqui tenemos en cuenta que se piden los elementos, no el indice, por lo tanto el resultado será: "'pop', 'rock'"

#**Encuentra el primer índice de <code>"disco"</code>**

```python
genres_tuple = ("pop", "rock", "soul", "hard rock", "soft rock",
                "R&B", "progressive rock", "disco")

indice = genres_tuple.index("disco")

print({indice})
```

**Respuesta:** Se hizo uso de "genres_tuple.index("disco")", ya que nos ayuda a conocer el indice teniendo en cuenta la informacion de la tupla, por lo tanto el indice sera 7

#**Genera una lista ordenada desde la tupla <code>C_tuple=(-5, 1, -3)</code>**

```python
C_tuple = (-5, 1, -3)
lista_ordenada = sorted(C_tuple)
print(lista_ordenada)
```

**Respuesta:** Se hizo uso de "sorted(C_tuple)", para ordenar la tupla dada en la pregunta, lo que nos da un resultado de: "[-5, -3, 1]"
