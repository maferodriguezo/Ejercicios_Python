#**Cree una función que divida la primera entrada por la segunda entrada**

```python
def dividir (a,b):
  return a/b

print(dividir(10,2))
```

**Respuesta:** Se usaron "a" y "b", como entradas, y re retona su division, una linea más abajo se usa en ejemplo donde se divide 10 entre 2, dando como resultado "5.0" garantizando su correcta ejecucion

**Utilice la función <code> con </code> para la siguiente pregunta.** *texto en cursiva*

```python
def con(a, b):
    return(a + b)
```

#**¿Se puede usar la función <code> con </code> que definimos antes para agregar números enteros o cadenas?**

```python
print(con(3, 5))  #Suma de números

print(con("Hola", " Mundo"))  # Output: Suma de cadenas
```

**Respuesta:** Si se puede, al ejecutar el primer print se hara la suma de 3 y 5, que son los parametros enteros que le hemos dado, lo cual da un resultado de "8".

Por otro lado hacer uso de cadenas tambien es posible, hemos usado "Hola" y " Mundo", lo cual dara como resultado la mezcla de estos dos elementos por lo tanto "Hola Mundo"

#**¿Se puede usar la función <code> con </code> que definimos antes para concentrar una lista o tupla?**

```python
print(con([1, 2], [3, 4]))  # Lista

print(con((1, 2), (3, 4)))  # Tupla
```

**Respuesta:** Si se puede, se hizo una prueba con "print(con([1, 2], [3, 4]))" y "print(con((1, 2), (3, 4)))", y funcionan en ambos. Pero deben tener elementos del mismo, si hubiera lista y entero por ejemplo, no se podria usar la funcion, asi mismo si se mezclan las tuplas con las listas en una sola
