#**Solución de cuestionario sobre conjuntos**

#**Transformar la lista <code>['rap','house','electronic music', 'rap']</code> en un conjunto:**

```python
lista_musica = ['rap', 'house', 'electronic music', 'rap']
conjunto_musica = set(lista_musica)
print(conjunto_musica)
```

**Respuesta:** Se hizo uso de "set(lista_musica)", lo cual nos ayudo a crear un conjunto basado en una lista

#**Considera que la lista <code>A = [1, 2, 2, 1]</code> y el conjunto <code>B = set([1, 2, 2, 1])</code> suman <code>sum(A) = sum(B)</code>**

```python
A = [1, 2, 2, 1]       # Lista con elementos repetidos
B = set([1, 2, 2, 1])  # Conjunto (debe eliminar duplicados)

suma_A = sum(A)

suma_B = sum(B)

print(suma_A == suma_B)
```

**Respuesta:** Se hizo uso de "sum(A) y sum(B)", pero estas sumas no son equivalentes, ya que no se esta  teniendo en cuenta que un conjunto no tiene elementos duplicados , en cambio la lista si, por eso mismo, no son equivalentes

#**Crea un nuevo conjunto <code>album_set3</code> que sea la unión de <code>album_set1</code> y <code>album_set2</code>:**

```python
album_set1 = set(["Thriller", 'AC/DC', 'Back in Black'])
album_set2 = set(["AC/DC", "Back in Black", "The Dark Side of the Moon"])

# Unión de conjuntos
album_set3 = album_set1.union(album_set2)
print(album_set3)
```

**Respuesta:** Se hizo uso de "album_set1.union(album_set2)", para unir esos dos conjuntos y se guardo en un nuevo conjunto llamado "album_set3" y se ve asi: "{'Back in Black', 'AC/DC', 'The Dark Side of the Moon', 'Thriller'}"

#**Averigua si <code>album_set1</code> es un subconjunto de <code>album_set3</code>**

```python
subconjunto = album_set1.issubset(album_set3)
print(subconjunto)
```

**Respuesta:** Se hizo uso de "issubset", para encontrar si el conjunto 1 y es subconjunto del 3, lo cual nos da una salida de verdadero
