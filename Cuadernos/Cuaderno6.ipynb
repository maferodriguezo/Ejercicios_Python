#**Escribe una declaración if que determine si un disco musical tiene una calificación mayor a 8. Compruébalo utilizando el calificación del disco <b>“Back in Black”</b> el cual tiene un valor de 8.5. Si la condición es verdadera, imprime “"This album is Amazing!"”**

```python
# Definimos la calificación de "Back in Black"
Back_in_back = 8.5

if Back_in_back > 8:
    print("This album is Amazing!")
```

**Respuesta:** Se hizo uso de "if > 8", lo cual nos ayudo comparar la calificación del del disco con el item que nos daban, por lo tanto en pantalla se vera un "This album is Amazing!"

#**Escribe una declaración if-else que determine lo siguiente. Si la calificación es mayor que 8 imprime “this album is amazing”. Si es menor o igual a 8 que imprima “this album is ok”.**

```python
album = 5

if album > 8:
    print("This album is Amazing")

elif album <= 8:
    print("This album is ok")
```

**Respuesta:** Se hizo uso de "if > 8" y "elif <= 8", lo cual nos ayudo comparar la calificación del del disco teniendo en cuenta dos conficiones. En este caso se uso una puntuacion de 5 para comporbar que funcionara de manera correcta, por lo tanto el mensaje de salida es "This album is ok"

#**Escribe una declaración if que determine si un disco fue lanzado antes de 1980 o en los años 1991 o 1993. Si la condición es verdadera imprime el año de lanzamiento.**

```python
lanzamiento = 1991

if lanzamiento < 1980 or lanzamiento in [1991, 1993]:
    print(f"El disco fue lanzado en el año: {lanzamiento}")
```

**Respuesta:** Se hizo uso de "or" e "in", ya que antes del "or" buscana numeros menores y la otra condicion que tambien podia ocurrir tenia dos valores puntuales, por eso el "in". Se hizo un ejemplo donde se cumple por lo tanto se muestra en pantalla el año en el que se lanzo el disco: "El disco fue lanzado en el año: 1991"
