# **¿Cual es el valor de la variable <code>A</code> después de ejecutar el siguiente código?**

```python
A = "1"
```

**Respuesta:** El valor de A será una cadena conformada por un único carácter: el "1". Esto se debe a que en Python, cualquier valor entre comillas (simples o dobles) se interpreta como cadena, incluso si contiene dígitos numéricos.

#**¿Cual es el valor de la variable <code>B</code> después de ejecutar el siguiente código?**

```python
B = "2"
```

**Respuesta:** El valor de B será una cadena conformada por un solo carácter: el "2". Esto se debe a que en Python, cualquier valor entre comillas (simples o dobles) se interpreta como cadena, incluso si contiene dígitos numéricos, al igual que el punto anterior.

#**¿Cual es el valor de la variable <code>C</code> después de ejecutar el siguiente código?**

```python
C = A + B

print(C)
```

**Respuesta:** Se va a ver una cadena de dos dígitos "12" tal cual sin espacios, ya que esta uniendo las cadenas de los puntos anteriores

#**En la variable <code>D</code> haz uso del slicing para imprimir solo los primeros tres elementos:**

```python
D = "ABCDEFG"

#Se crea una nueva variable que guarde los datos obtenidos al hacer "Slicing" en la cadena D

Primeros_tres = D[0:3]

print(Primeros_tres)
```

**Respuesta:** Al compilar se muestra en pantalla los tres primeros digitos de la cadena "ABC"

#**Usa un valor de stride de 2 para imprimir cada segundo elemento de la cadena <code>E</code>:**

```python
E = 'clocrkr1e1c1t'

#Se crea una nueva variable que guarde los datos obtenidos al hacer "Stride" en la cadena E

Stride = E[::2]

print(Stride)
```

**Respuesta:** Al compilar se muestra en pantalla cada segundo elemento en la cadena E "correct"
