#**Crea una lista <code>a_lit</code> con los siguientes elementos <code>1</code>, <code>hello</code>, <code>[1,2,3]</code> y <code>True</code>**

```python
a_list = [1, "hello", [1, 2, 3], True]

print(a_list)
```

**Respuesta:** Se hizo uso de "a_list = [1, "hello", [1, 2, 3], True]", lo cual nos creo la lista pedida

#**Encuentra el valor almacenado en el índice 1 de <code>a_list</code>.**

```python
indice_1 = a_list[1]  # Accede al índice 1
print(indice_1)       # Imprime el valor
```

**Respuesta:** Se hizo uso de "a_list[1]", para encontrar el elemento del indice 1, el cual es "hello"

#**Obtén el elemento almacenado en el índice 1, 2 y 3 de <code>a_list</code>.**

```python
Slicing = a_list[1:4]
print(Slicing)
```

**Respuesta:** Se hizo uso de "Slicing", para los elementos de los indices del 1 al 3, los cuales son "'hello', [1, 2, 3], True"

#**Concatena las siguientes listas <code>A = [1, 'a']</code> y <code>B = [2, 1, 'd']</code>:**

```python
A = [1, 'a']
B = [2, 1, 'd']
concatenada = A + B
print(concatenada)
```

**Respuesta:** Se hizo uso de "+", ya que este es usando para concatenar, como resultado veremos una nueva lista "[1, 'a', 2, 1, 'd']"
